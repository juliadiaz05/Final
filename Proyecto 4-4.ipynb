## Punto de partida: Proyecto 2

Importamos las librerias con las que vamos a trabajar

```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
```

**Importo el Dataset y veo sus datos**

```python
proyecto = pd.read_csv('DS_Proyecto_01_Datos_Properati.csv') #importamos el dataset 
proyecto.head(2)
```

**Voy a realizar un filtro sobre propiedad, trabajando sólo con departamento, PH y casa de Capital Federal; la superficie total debe ser mayor a la cubierta y debe tener como máximo 5 baños. El valor máximo a tomar por propiedad es de hasta USD$400.000.**

```python
nprop = proyecto.property_type.isin(['Departamento', 'Casa', 'PH'])
nsup=(proyecto['surface_total']>=proyecto['surface_covered'])
nprecio= proyecto['price']<=400000
ncap= proyecto['l2']=='Capital Federal'
nban= proyecto['bathrooms']<=5


proyecto=proyecto[nprop & nsup & nprecio & ncap & nban]
proyecto.shape
```

**Realizamos Encoding con la variable "property_type"**

```python
dummies= pd.get_dummies(proyecto['property_type'], prefix = 'property_type')
dummies
```

**Concatenamos el Encoding con el Dataset**

```python
concatenado_uno =pd.concat([proyecto,dummies], axis=1)
concatenado_uno.head(2)
```

**Elimino variables que no utilizo**

```python
proyecto_uno = concatenado_uno.drop(columns = ['start_date', 'end_date', 'currency','title', 'description', 'operation_type', 'l1', 'created_on', 'property_type', 'l2','l3'])
proyecto_uno.head(2)
```

```python
proyecto_uno.dropna(inplace= True)
```

```python
proyecto_uno.isna().sum()
```

```python
proyecto_uno.shape
```

**Detección y remoción de outliers por el criterio de tres sigmas**

```python
from scipy.stats import zscore

zscore_precio = zscore(proyecto_uno['price'])
zscore_suptot = zscore(proyecto_uno['surface_total'])
zscore_supcub = zscore(proyecto_uno['surface_covered'])
filtro_zscore = (abs(zscore_precio)>3) |  (abs(zscore_suptot)>3) | (abs(zscore_supcub)>3)
proyecto_dos= proyecto_uno[~filtro_zscore]
```

```python
proyecto_dos.shape
```

**Como modelo Benchmark voy a realizar Random Forest sobre el modelo sin los valores de Latitud y Longitud y luego los voy a comparar con el resultado del modelo de boosting con esas variables incluidas**

**Definimos X e Y, realizamos train/test split sobre el modelo benchmark**

```python
X_uno = proyecto_dos[['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'property_type_Casa', 'property_type_Departamento', 'property_type_PH']] 
y = proyecto_dos['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_uno, y, test_size=0.2, random_state=42)
```

**Modelos Avanzados**

Vamos a utilizar Grid Search, Random Forest y XGBoost

Redefinimos nuevamente las variables


```python
def entrena_modelo (X_train, X_test, y_train, y_test, modelo):
    
    #Entrenamos el modelo y predecimos y_train e y_test:
    modelo.fit(X_train, y_train)
    y_train_pred = modelo.predict(X_train)
    y_test_pred = modelo.predict(X_test)
    
    #Imprimimos las metricas consideradas para evaluar el modelo:
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    rf = modelo.score(X_test, y_test)
    print(f'Score de modelo: {rf}')
    print(f'Raiz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raiz del error cuadrático medio en Test: {rmse_test}')
    
    #Graficamos la distribucion de errores:
    plt.figure(figsize = (15,6))
    plt.subplot(1,2,1)
    sns.distplot(y_train - y_train_pred, bins = 20, label = 'train')
    sns.distplot(y_test - y_test_pred, bins = 20, label = 'test')
    plt.xlabel('errores')
    plt.legend()

    ax = plt.subplot(1,2,2)
    plt.scatter(y_test,y_test_pred, s =2)    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()
    plt.show()
    

    return rmse_train,rmse_test,rf
```

**GridSearch**

```python
n_estimators = [100,250,500]
max_depth = [10,15,30]


grid= {'n_estimators': n_estimators,
               'max_depth': max_depth,}
```

```python
modelorf = GridSearchCV(RandomForestRegressor(random_state = 42),param_grid = grid, n_jobs = -1, cv = 3)
```

```python
rmse_train_rf,rmse_test_rf,score_rf = entrena_modelo(X_train, X_test, y_train, y_test, modelorf)
```

**XGBoost**

**Definimos nuevamente X con las variables de LON y LAT incluidas.**

```python
X_ll = proyecto_dos[['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'property_type_Casa', 'property_type_Departamento', 'property_type_PH', 'lat', 'lon']]
```

```python
X_train_ll, X_test_ll, y_train, y_test = train_test_split(X_ll, y, test_size=0.2, random_state=42)
```

```python
def entrena_modelo (X_train_ll, X_test_ll, y_train, y_test, modelo):
    
    #Entrenamos el modelo y predecimos y_train e y_test:
    modelo.fit(X_train_ll, y_train)
    y_train_pred_ll = modelo.predict(X_train_ll)
    y_test_pred_ll = modelo.predict(X_test_ll)
    
    #Imprimimos las metricas consideradas para evaluar el modelo:
    rmse_train_ll = np.sqrt(mean_squared_error(y_train, y_train_pred_ll))
    rmse_test_ll = np.sqrt(mean_squared_error(y_test, y_test_pred_ll))
    rf_ll = modelo.score(X_test_ll, y_test)
    print(f'Score de modelo: {rf_ll}')
    print(f'Raiz del error cuadrático medio en Train: {rmse_train_ll}')
    print(f'Raiz del error cuadrático medio en Test: {rmse_test_ll}')
    
    #Graficamos la distribucion de errores:
    plt.figure(figsize = (15,6))
    plt.subplot(1,2,1)
    sns.distplot(y_train - y_train_pred_ll, bins = 20, label = 'train')
    sns.distplot(y_test - y_test_pred_ll, bins = 20, label = 'test')
    plt.xlabel('errores')
    plt.legend()

    ax = plt.subplot(1,2,2)
    plt.scatter(y_test,y_test_pred_ll, s =2)    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()
    plt.show()
    
    return rmse_train_ll,rmse_test_ll,rf_ll
```

```python
eta = [0,1]
max_depth = [10,15,20]

```

```python
grid= {'learning_rate': eta,
               'max_depth': max_depth,}
```

```python
modelxgb = GridSearchCV(XGBRegressor(random_state=42), param_grid =grid, n_jobs = -1)
```

```python
rmse_train_xgb,rmse_test_xgb,scorexgb = entrena_modelo(X_train_ll, X_test_ll, y_train, y_test, modelxgb)
```

```python
best_params =  modelxgb.best_params_
print (f'Los mejores parametros son: {best_params}')
```
